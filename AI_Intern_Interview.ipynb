```python
import os
import shutil

ds_store_path = os.path.join("trashnet_data/dataset-resized", ".DS_Store")
if os.path.exists(ds_store_path):
    os.remove(ds_store_path)

os.makedirs("data_images", exist_ok=True)
categories = ['glass', 'trash', 'metal', 'plastic', 'cardboard', 'paper']
for category in categories:
    src = os.path.join("trashnet_data/dataset-resized", category)
    dst = os.path.join("data_images", category)
    shutil.copytree(src, dst)

shutil.rmtree("trashnet_data", ignore_errors=True)

print("Remaining directories in 'data_images':", os.listdir("data_images"))
```

```python
import torch
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])


dataset = datasets.ImageFolder("data_images", transform=transform)

# Split the dataset (80%, 20%)
total_size = len(dataset)
train_size = int(0.8 * total_size)
val_size = total_size - train_size

train_dataset, val_dataset = random_split(
    dataset, [train_size, val_size],
    generator=torch.Generator().manual_seed(42)
)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

print("Classes:", dataset.classes)
print("Train size:", len(train_dataset))
print("Validation size:", len(val_dataset))
```

<h3> Convolution and Pooling formulas: </h3>

$$
n_{out} = \left\lfloor \frac{n_{in} + 2p - k}{s} \right\rfloor + 1
$$

Where:  
- n_in: number of input features  
- n_out: number of output features  
- k: kernel size  
- p: padding  
- s: stride

<hr>

<h3> Transpose Convolution formula: </h3>
$$
n_{out} = (n_{in} - 1) \cdot s - 2p + k
$$

Where:  
- n_in: number of input features  
- n_out: number of output features  
- k: kernel size  
- p: padding  
- s: stride
